```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import class_weight

np.random.seed(42)
tf.random.set_seed(42)
```

```python
def load_and_preprocess_data(base_path, target_size=(150, 150)):
    images = []
    masks = []
    labels = []
    categories = ['benign', 'malignant', 'normal']
    
    for idx, category in enumerate(categories):
        path = os.path.join(base_path, category)
        print(f"Processing {category} images...")
        
        # Get all image files first
        image_files = [f for f in os.listdir(path) if not f.endswith('_mask.png') and f.endswith('.png')]
        
        # Add data augmentation for normal class to balance dataset
        augment_factor = 3 if category == 'normal' else 1
        
        for img_name in image_files:
            img_path = os.path.join(path, img_name)
            mask_path = os.path.join(path, img_name.replace('.png', '_mask.png'))
            
            # Load original image
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            img = cv2.equalizeHist(img)
            img = cv2.resize(img, target_size)
            
            # Load mask if exists
            if os.path.exists(mask_path):
                mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
                mask = cv2.resize(mask, target_size)
            else:
                mask = np.zeros(target_size, dtype=np.uint8)
            
            # Normalize images
            img_normalized = img / 255.0
            img_normalized = np.expand_dims(img_normalized, axis=-1)
            
            # Add original image
            images.append(img_normalized)
            masks.append(mask)
            labels.append(idx)
            
            # Add augmented versions for normal class
            if category == 'normal':
                for _ in range(augment_factor - 1):
                    # Apply random augmentations
                    angle = np.random.randint(-20, 20)
                    scale = 1.0 + np.random.uniform(-0.1, 0.1)
                    
                    M = cv2.getRotationMatrix2D((target_size[0]//2, target_size[1]//2), angle, scale)
                    
                    aug_img = cv2.warpAffine(img, M, target_size)
                    aug_img = aug_img / 255.0
                    aug_img = np.expand_dims(aug_img, axis=-1)
                    
                    aug_mask = cv2.warpAffine(mask, M, target_size)
                    
                    images.append(aug_img)
                    masks.append(aug_mask)
                    labels.append(idx)
    
    return np.array(images), np.array(masks), np.array(labels), categories
```

```python
def create_model(input_shape=(150, 150, 1)):
    model = Sequential([
        # First Block
        Conv2D(64, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Second Block
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Third Block
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        Conv2D(256, (3, 3), activation='relu', padding='same'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        Dropout(0.25),
        
        # Dense Layers
        Flatten(),
        Dense(512, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(256, activation='relu'),
        BatchNormalization(),
        Dropout(0.5),
        Dense(3, activation='softmax')
    ])
    
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model
```

```python
def visualize_prediction(image_path, model, categories, target_size=(150, 150)):
    # Load and preprocess image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    original_img = img.copy()
    
    # Load mask if exists
    mask_path = image_path.replace('.png', '_mask.png')
    if os.path.exists(mask_path):
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    else:
        mask = np.zeros_like(img)
    
    # Preprocess for prediction
    img = cv2.equalizeHist(img)
    img = cv2.resize(img, target_size)
    img_normalized = img / 255.0
    img_normalized = np.expand_dims(img_normalized, axis=-1)
    img_normalized = np.expand_dims(img_normalized, axis=0)
    
    # Make prediction
    prediction = model.predict(img_normalized)
    predicted_class = categories[np.argmax(prediction)]
    confidence = float(np.max(prediction)) * 100
    
    # Visualization
    plt.figure(figsize=(15, 5))
    
    # Original image
    plt.subplot(131)
    plt.imshow(original_img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    # Mask overlay
    plt.subplot(132)
    mask_overlay = cv2.resize(mask, target_size)
    plt.imshow(img, cmap='gray')
    plt.imshow(mask_overlay, cmap='hot', alpha=0.3)
    plt.title('Mask Overlay')
    plt.axis('off')
    
    # Prediction probabilities
    plt.subplot(133)
    bars = plt.bar(categories, prediction[0] * 100)
    plt.title(f'Prediction: {predicted_class}\nConfidence: {confidence:.2f}%')
    plt.ylabel('Probability (%)')
    plt.xticks(rotation=45)
    
    # Color the bars
    for idx, bar in enumerate(bars):
        if idx == np.argmax(prediction):
            bar.set_color('green')
        else:
            bar.set_color('gray')
    
    plt.tight_layout()
    plt.show()
    
    return predicted_class, confidence
```

```python
# Main execution
if __name__ == "__main__":
    base_path = "Dataset_BUSI_with_GT"
    
    # Load data with masks
    print("Loading and preprocessing data...")
    images, masks, labels, categories = load_and_preprocess_data(base_path)
    print(f"Total images: {len(images)}")
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        images, labels, test_size=0.2, random_state=42, stratify=labels
    )
    
    # Calculate class weights with higher weight for normal class
    class_weights = class_weight.compute_class_weight(
        'balanced',
        classes=np.unique(labels),
        y=labels
    )
    # Increase weight for normal class
    class_weights[-1] *= 1.5
    class_weights = dict(enumerate(class_weights))
    print("\nClass weights:", class_weights)
    
    # Train model
    model = create_model()
    model.summary()
    
    callbacks = [
        EarlyStopping(patience=15, restore_best_weights=True),
        ReduceLROnPlateau(factor=0.2, patience=7, min_lr=1e-6)
    ]
    
    print("\nTraining model...")
    history = model.fit(
        X_train, y_train,
        batch_size=32,
        epochs=100,
        validation_data=(X_test, y_test),
        class_weight=class_weights,
        callbacks=callbacks
    )
    
    # Evaluate model
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_classes, target_names=categories))
    
    # Plot confusion matrix
    plt.figure(figsize=(10, 8))
    cm = confusion_matrix(y_test, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=categories, yticklabels=categories)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    # Plot training history
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
```

```python
def check_mask(image_path):
    mask_path = image_path.replace('.png', '_mask.png')
    has_mask = os.path.exists(mask_path)
    
    if has_mask:
        mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
        mask_exists = np.any(mask > 0)  # Check if mask contains any non-zero pixels
        return has_mask, mask, mask_exists
    return False, np.zeros((150, 150), dtype=np.uint8), False

def preprocess_image(image_path, target_size=(150, 150)):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    original_img = img.copy()
    
    # Preprocess
    img = cv2.equalizeHist(img)
    img = cv2.resize(img, target_size)
    img_normalized = img / 255.0
    img_normalized = np.expand_dims(img_normalized, axis=-1)
    img_normalized = np.expand_dims(img_normalized, axis=0)
    
    return img_normalized, original_img

def get_prediction(model, img_normalized, categories):
    prediction = model.predict(img_normalized)
    probabilities = prediction[0]
    class_indices = np.argsort(probabilities)[::-1]  
    
    return probabilities, class_indices

def adjust_prediction_based_on_mask(probabilities, class_indices, categories, has_mask, mask_exists):
    normal_idx = categories.index('normal')
    
    if not mask_exists:
        predicted_class = 'normal'
        confidence = 100.0
        predicted_idx = normal_idx
        print("Note: Classified as normal due to absence of mask.")
    else:
        predicted_idx = class_indices[0]
        if categories[predicted_idx] == 'normal':
            predicted_idx = class_indices[1]
            print("Note: Using second highest prediction due to presence of mask.")
        
        predicted_class = categories[predicted_idx]
        confidence = float(probabilities[predicted_idx]) * 100
    
    return predicted_class, confidence, predicted_idx

def visualize_results(original_img, mask, probabilities, categories, predicted_class, confidence, mask_exists, target_size=(150, 150)):
    plt.figure(figsize=(15, 5))
    
    # Original image
    plt.subplot(131)
    plt.imshow(original_img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    
    # Mask overlay
    plt.subplot(132)
    mask_overlay = cv2.resize(mask, target_size)
    img_resized = cv2.resize(original_img, target_size)
    plt.imshow(img_resized, cmap='gray')
    if mask_exists:
        plt.imshow(mask_overlay, cmap='hot', alpha=0.3)
        plt.title('Mask Overlay')
    else:
        plt.title('No Mask Found - Normal Case')
    plt.axis('off')
    
    # Prediction probabilities
    plt.subplot(133)
    if not mask_exists:
        modified_probs = np.zeros_like(probabilities)
        modified_probs[categories.index('normal')] = 1.0
        bars = plt.bar(categories, modified_probs * 100)
    else:
        bars = plt.bar(categories, probabilities * 100)
    
    plt.title(f'Prediction: {predicted_class}\nConfidence: {confidence:.2f}%')
    plt.ylabel('Probability (%)')
    plt.xticks(rotation=45)
    
    # Color the bars
    for idx, bar in enumerate(bars):
        if categories[idx] == predicted_class:
            bar.set_color('green')
        else:
            bar.set_color('gray')
    
    plt.tight_layout()
    plt.show()

def predict_and_visualize(image_path, model, categories, target_size=(150, 150)):
    # Check for mask
    has_mask, mask, mask_exists = check_mask(image_path)
    
    # Preprocess image
    img_normalized, original_img = preprocess_image(image_path, target_size)
    
    # Get initial prediction
    probabilities, class_indices = get_prediction(model, img_normalized, categories)
    
    # Adjust prediction based on mask
    predicted_class, confidence, predicted_idx = adjust_prediction_based_on_mask(
        probabilities, class_indices, categories, has_mask, mask_exists
    )
    
    # Visualize results
    visualize_results(original_img, mask, probabilities, categories, 
                     predicted_class, confidence, mask_exists, target_size)
    
    # Print detailed probabilities
    print("\nDetailed Probabilities:")
    if not mask_exists:
        print("No mask found - Automatically classified as normal (100%)")
    else:
        for cat, prob in zip(categories, probabilities):
            print(f"{cat}: {prob*100:.2f}%")
    
    return predicted_class, confidence

def process_multiple_images(image_paths, model, categories):
    results = []
    for image_path in image_paths:
        if os.path.exists(image_path):
            predicted_class, confidence = predict_and_visualize(image_path, model, categories)
            results.append({
                'image': image_path,
                'prediction': predicted_class,
                'confidence': confidence,
            })
            print(f"\nProcessed: {image_path}")
            print(f"Final Prediction: {predicted_class}")
            print(f"Confidence: {confidence:.2f}%")
            print("-" * 50)
    return results

# Example 
if __name__ == "__main__":
    # Load your saved model
    model = tf.keras.models.load_model('breast_ultrasound_model.h5')
    categories = ['benign', 'malignant', 'normal']
    base_path = "Dataset_BUSI_with_GT"
    
    # Test single image
    test_image_path = os.path.join(base_path, "malignant/malignant (1).png")
    if os.path.exists(test_image_path):
        predicted_class, confidence = predict_and_visualize(test_image_path, model, categories)
        print(f"\nFinal Prediction:")
        print(f"Class: {predicted_class}")
        print(f"Confidence: {confidence:.2f}%")

    # Test multiple images
    test_images = [
        os.path.join(base_path, "benign/benign (55).png"),
        os.path.join(base_path, "malignant/malignant (55).png"),
        os.path.join(base_path, "normal/normal (55).png")
    ]
    results = process_multiple_images(test_images, model, categories)
```
